4. Utilizando el conjunto de datos Iris, aplica el método de mínimos cuadrados para encontrar el mejor ajuste lineal que separe las dos clases de flores.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Cargar el conjunto de datos Iris
iris = datasets.load_iris()
X = iris.data[:, :2]  # Tomar solo las primeras dos características
y = iris.target

# Seleccionar solo las dos primeras clases de iris
X = X[y != 2]
y = y[y != 2]

# Agregar una columna de unos para el término independiente
X = np.c_[np.ones(X.shape[0]), X]

# Dividir los datos en dos clases
X0 = X[y == 0]
X1 = X[y == 1]

# Aplicar mínimos cuadrados para encontrar la línea de separación
w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# Graficar los datos y la línea de separación
plt.figure()
plt.scatter(X0[:, 1], X0[:, 2], color='blue', label='Clase 0')
plt.scatter(X1[:, 1], X1[:, 2], color='red', label='Clase 1')
plt.plot([4, 7], [-w[0]/w[2] - w[1]/w[2]*4, -w[0]/w[2] - w[1]/w[2]*7], color='black', label='Separación')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Ancho del sépalo')
plt.legend()
plt.show()
```
